```python
# 결측값으로 추가. 원산지 컬럼을 만들고 결측값(NAN)으로 대입
import numpy as np

df['원산지'] = np.nan
df
```

```python
  # 리스트 형태로 데이터(행) 추가(loc나 iloc 사용해야함) df.loc[인덱스명

affel=['아이펠치킨',16000,'11호',1200, 0.5, 8000, '국내산']

df.loc['new']=affel
df
```

```python
  # 딕셔너리 형태로 데이터(행) 추가

flip={'메뉴':'풀잎치킨', '가격':10000, '호수':'10호'}

df.loc[10]=flip
df
```

```python
#후라이드 원산지는 '국내산' 값 추가(loc활용)

df.loc[0,'원산지']='국내산'
df
```

```python
   # 파닭과 닭강정의 원산지는 '브라질' 값 추가 (iloc 사용)

df.iloc[4,-1]='브라질'
df.iloc[5,6]='브라질'
df
```

```python
# replace활용(양반후반->양념반후라이드반)

df= df.replace('양반후반','양념반 후라이드반')
df
```

```python
# replace활용(아이펠치킨->[인기]아이펠치킨,풀잎치킨->[베스트]풀잎치킨)

df=df.replace('아이펠치킨','[인기]아이펠치킨').replace('풀잎치킨','[베스트]풀잎치킨')
df


```

```python
   # replace, 딕셔너리 활용(후라이드->황금후라이드, 양념치킨->승일양념치킨)

d= {'후라이드':'황금후라이드','양념치킨':'승일양념치킨'}
df= df.replace(d)
df
```

```python
   # 숫자변경 : 10000->9900
df =df.replace(10000,9900)
df
```

```python
df.to_csv("new_modudak.csv")
df
```
