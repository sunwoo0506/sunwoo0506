```python
# 1) 리스트형태로 넣기
df = data[['가격','호수','칼로리','원산지','살찔까요']].copy()
df

from sklearn.preprocessing import OneHotEncoder

ohe=OneHotEncoder(sparse=False)

cat = ohe.fit_transform(df[['원산지','살찔까요']])
cat

df_cat = pd.DataFrame(cat, columns = ohe.get_feature_names_out())

df_cat

df = pd.concat([df,df_cat], axis=1)
df

df = df.drop(['원산지','살찔까요'], axis=1)
df
```

```python
# 2) select_dtypes 사용

df = data[['가격','호수','칼로리','원산지','살찔까요']].copy()
df

from sklearn.preprocessing import OneHotEncoder

ohe=OneHotEncoder(sparse=False)
cols = df.select_dtypes(include='object').columns
cat = ohe.fit_transform(df[cols])

df_cat = pd.DataFrame(cat,columns=ohe.get_feature_names_out())
df = pd.concat([df,df_cat], axis=1)
df

df = df.drop(cols, axis=1)
df
```

```python
# 4. 사이킷런에서 제공하는 데이터셋
# 유방암 데이터

from sklearn.datasets import load_breast_cancer
dataset = load_breast_cancer()
dataset
```

```python
dataset.feature_names
```

```python
dataset.target
```

```python
dataset.data[:2]
```

```python
cancer_df = pd.DataFrame(data=dataset.data, columns=dataset.feature_names)
cancer_df
```

```python
cancer_df['target'] = dataset.target
cancer_df.head()
```

```python
# 당뇨병데이터 불러오기
from sklearn.datasets import load_diabetes

dataset = load_diabetes()
dataset

dataset.feature_names
dataset.target
dataset.data[:2]

diabetes_df=pd.DataFrame(data=dataset.data, columns=dataset.feature_names)
diabetes_df

diabetes_df['target']=dataset.target
diabetes_df.head()
```

```python
cancer_df.head()
```

```python
# 머신러닝 : 분류

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    cancer_df.drop('target', axis=1), cancer_df['target'], test_size=0.3, random_state=1004)

X_train.head()
```

```python
y_train.head()
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# 의사결정나무(DecisionTree)

from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(X_train,y_train)
pred = model.predict(X_test)
pred
```

```python
#정확도 accuracy_score(실제값,예측값)

from sklearn.metrics import accuracy_score
accuracy_score(y_test,pred)
```

```python
#머신러닝: 회귀

#검증데이터 분리

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(diabetes_df.drop('target',axis=1), diabetes_df['target'], test_size=0.3, random_state=1004)

X_train.head()
```

```python
y_train.head()
```

```python
# 데이터의 크기

X_train.shape, X_test.shape, y_train.shape, y_test.shape

```

```python
# 모델 선택

from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X_train,y_train)

pred= model.predict(X_test)
pred
```

```python
#평가(MSE, mean_squared_error(실제값,예측))

from sklearn.metrics import mean_squared_error
mean_squared_error(y_test,pred)

```
